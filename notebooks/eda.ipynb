```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import pickle
```

```python
with open('../data/processed/blogcatalog.gpickle', 'rb') as f:
    G = pickle.load(f)

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

```python
degrees = [d for n, d in G.degree()]
plt.figure(figsize=(8,4))
plt.hist(degrees, bins=50, color='steelblue')
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('../results/figures/degree_distribution.png', dpi=300)
plt.show()
```

```python
labels_df = pd.read_csv('../data/processed/labels_blogcatalog.csv')
print(labels_df.head())
```

```python
label_counts = labels_df.drop('node_id', axis=1).sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
label_counts.plot(kind='bar', color='coral')
plt.title("Label Frequency Distribution")
plt.xlabel("Label")
plt.ylabel("Number of Nodes")
plt.tight_layout()
plt.savefig('../results/figures/label_frequency.png', dpi=300)
plt.show()
```

```python
node_counts = labels_df.drop('node_id', axis=1).sum()

plt.figure(figsize=(12,5))
node_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Node Count Per Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('../results/figures/node_count_per_class.png', dpi=300)
plt.show()
```
